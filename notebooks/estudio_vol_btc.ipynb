```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickers = ["BTC-USD", "^GSPC", "GC=F"]
start_date = "2018-01-01"
end_date = None

data = yf.download(tickers, start=start_date, end=end_date, interval = "1d", progress = False)

close = data["Close"]

ret_log = np.log(close / close.shift(1))

vol_anual = ret_log.groupby(ret_log.index.year).std() * np.sqrt(252)

vol_anual.plot(kind='bar', figsize = (10, 6))


```
